```python
import numpy as np
import math as m
import scipy as scp
from scipy.integrate import odeint
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import matplotlib.collections as coll
import scipy
import time
import cplex
import operator
import networkx as nx
```

```python
r = 0.025 #radius of wheels
L = 0.09 # length of axle
W = 0.1 # width of robot
#r = 0.5
#L = 1.
#W = 1.

op_space_xmin = 0.
op_space_xmax = 1.
op_space_ymin = 0.
op_space_ymax = 1.

limits = np.array([4.*np.pi, 4.*np.pi])

obstacles = 0.1*np.array([[4, 5, 4, 5],
                          [2, 3, 2, 3],
                          [4, 5, 0, 1]
                       ])

def dynamics(x,t,u):
    global r
    global L
    xdot = r*np.cos(x[2])*(u[0]+u[1])/2.0
    ydot = r*np.sin(x[2])*(u[0]+u[1])/2.0
    theta_dot = r/L*(u[1]-u[0])
    return np.array([xdot,ydot,theta_dot])
    

def delta_angle(x_1,x_2,current_theta):
    vect = (x_2-x_1)/(np.linalg.norm(x_2-x_1)) # normalized vector pointing between points
    current_orient = np.array([np.cos(current_theta), np.sin(current_theta)])
    sin_value = np.clip(np.cross(vect,current_orient),-1.,1.)
    cos_value = np.clip(np.dot(vect,current_orient),-1.,1.)
    return m.copysign(np.arcsin(sin_value),cos_value)

def distance_metric(x_1,x_2):
    trans = np.linalg.norm(x_2[0:2]-x_1[0:2])
    rot = abs(delta_angle(x_1[0:2],x_2[0:2],x_1[2]))
    return trans #+ 0.25*rot

def compute_input(state_1,state_2):
    global limits
    alpha = 10.0 # gain for trans
    beta = 3.0*alpha # gain for angle
    x_1 = state_1[0:2]
    x_2 = state_2[0:2]

    not_arrived = any(abs(x_2-x_1) >= 0.00001)
    
    #norm_diff = np.linalg.norm(x_2-x_1)
    norm_diff = not_arrived #make the distance metric an indicator for whether we have arrived
    u_trans = alpha*norm_diff #compute an amplified distance to set point
    #angle_diff = delta_angle(x_1,x_2,state_1[2])
    angle_diff = delta_angle(x_1,x_2,state_1[2])*not_arrived + (np.sign(state_2[2]-state_1[2]))*(not not_arrived) #difference in angle between heading and line connecting points
    u_trans = m.copysign(u_trans,angle_diff)
    
    u_rot = beta*angle_diff #compute an amplified distance between angles
    u = input_to_rpms(np.array([u_trans,u_rot]))
    
    u = np.clip(input_to_rpms(np.array([u_trans,u_rot])),-limits, limits)
    #u = np.copysign(limits,u)
    return u

def input_to_rpms(u):
    A = np.array([[0.5, 0.5],[-1, 1]])
    #rpms = scipy.optimize.lsq_linear(A,u,bounds=(-limits[1],limits[1]))['x']
    rpms = np.linalg.solve(A,u)
    #if any(abs(rpms) > limits[1]):
    #    over = np.argmax[abs(rpms)]
    #    rpms[over]
    #    rpms[(over+1)%2] = 
    return rpms

def compute_corners(state):
    global L
    global r
    global W
    x = state[0:2]
    theta = state[2]
    par = np.array([np.cos(theta), np.sin(theta)])
    perp = np.array([-np.sin(theta), np.cos(theta)])
    c1 = x + perp*L/2.0 + par*r
    c2 = c1 - perp*L
    c3 = c1 - par*W
    c4 = c2 - par*W
    return [c1,c2,c3,c4]

def in_obstacles(state):
    state_corners = compute_corners(state)
    global obstacles
    for obst in obstacles:
        for corner in state_corners:
            in_obst = (corner[0] >= obst[0])*(corner[0] <= obst[1])*(corner[1] >= obst[2])*(corner[1] <= obst[3])
            if in_obst:
                return True
    return False

def trajectory_clear(traj):
    for state in np.transpose(traj):
        if in_obstacles(state):
            return False
    return True
    
```

```python
time_windows = 6 # length of each set point command (s)
sub_steps = 1000 # number of closed loop control points between commands
sub_sub_integrations = 100 # number of points to integrate over between closed loop control points
setpoints = 2 #number of set point commands
t_sub_sub_integrations = np.linspace(0.,time_windows/sub_steps,sub_sub_integrations) # time holder for integrations
total_steps = (sub_steps)*setpoints+1 # total number of states to be saved


x_d = np.empty([3,setpoints]) #stores all set points
x_d = [[1,1,0],
       [0.0,1.0,0]]
       #[2,3,0]]

x = np.zeros([3,total_steps]) # states
u = np.zeros([2,total_steps]) # inputs
x[:,0] = np.array([0,0,0]) #initial state

for point in range(setpoints):
    for step in range(point*sub_steps,(point+1)*sub_steps):
        u[:,step] = compute_input(x[:,step],x_d[point])
        #u[:,step] = controller_try(x[:,step],x_d)
        sol = odeint(dynamics, x[:,step],t_sub_sub_integrations,args = (u[:,step],))
        x[:,step+1] = sol[-1,:]
#compute_input(x[:,0],x_d)   
```

```python
def t_sec_trajectory(x_0,x_d,t):
    steps = 300
    sub_steps = 200
    x = np.zeros([3,steps+1])
    u = np.zeros([2,steps+1])
    t = np.linspace(0.,t/steps,sub_steps)
    x[:,0] = x_0
    for step in range(steps):
        u[:,step] = compute_input(x[:,step],x_d)
        sol = odeint(dynamics,x[:,step],t,args=(u[:,step],))
        x[:,step+1] = sol[-1,:]
    return x
```

```python
x_0 = np.array([0,0,np.pi/4.0])
RRT = nx.Graph()
RRT.add_node(tuple(x_0))
num_vertices = 1000
step_lengths = 6.0
shift = np.array([op_space_xmin, op_space_ymin, 0])
scale = np.array([(op_space_xmax - op_space_xmin), (op_space_ymax-op_space_ymin), 2*np.pi])

goal = np.array([0.9,0.9,5*np.pi/4.])

goal_reached = False
for i in range(num_vertices):
    x_rand = np.multiply((np.random.rand(3) - shift),scale)
    while in_obstacles(x_rand): # re-draw if necessary
        x_rand = np.multiply((np.random.rand(3) - shift),scale)
    
    # Compute argmax from distance metric
    dists = dict.fromkeys(RRT.nodes)
    for v in RRT.nodes():
        dists[v] = distance_metric(x_rand,np.array(v))
    closest = min(dists.items(), key=operator.itemgetter(1))[0]
    #print(closest)
    # generate trajectory
    traj = t_sec_trajectory(np.array(closest),x_rand,step_lengths)
    #if trajectory is okay
    if trajectory_clear(traj) :
        x_new = traj[:,-1]
        if distance_metric(x_new,goal) <= 0.01:
            print('Reached goal!')
            goal_reached = True
        #print('Edge added! %s, neigbor: %s' %(x_new,closest))
        RRT.add_edge(tuple(x_new),closest)
        RRT[tuple(x_new)][closest].update(data=traj)
    else :
        i -= 1
    if goal_reached:
        break
```

```python
f = plt.figure(figsize=(20,20))
ax = plt.subplot(111)
rects = []
for node in RRT.nodes():
    ax.scatter(node[0],node[1], color ='blue')
    #ax.quiver(node[0],node[1],np.cos(node[2]),np.sin(node[2]),scale=50)

for obst in obstacles:
    rect = patches.Rectangle((obst[0],obst[2]), (obst[1]-obst[0]), (obst[3]-obst[2]))
    rects.append(rect)
ax.scatter(x_0[0],x_0[1],color='red',marker='o',s=500)
ax.scatter(goal[0],goal[1],color='green',marker='*',s=500)
obst_squares = coll.PatchCollection(rects,facecolor='black')
ax.add_collection(obst_squares)
for edge in RRT.edges():
    traj = RRT[edge[0]][edge[1]]['data']
    ax.plot(traj[0,:],traj[1,:],color='red',alpha=0.4)
    #ax.plot([edge[0][0],edge[1][0]],[edge[0][1],edge[1][1]], color='red',alpha=0.2)
plt.savefig('sample_curly_rrt.png')
plt.show()
```
